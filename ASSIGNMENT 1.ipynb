```python
import numpy as np
import pandas as pd
from numpy import random
```

CREATING DATAFRAME WITH 4 OBSERVATIONS AND 4 FEATURES

```python
data=pd.DataFrame({
    "Random value 1": random.randint(99,size=4),
    "Random value 2": random.randint(99,size=4),
    "Random value 3": random.randint(99,size=4),
    "Random value 4": random.randint(99,size=4),
})
data
```

```python
data.describe()
```

```python
data.isnull().any()
```
